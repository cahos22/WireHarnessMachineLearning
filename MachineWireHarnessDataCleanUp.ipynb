```python
import pandas as pd
import numpy as np
```

```python
sample_data = pd.read_csv('Data/OperationLog.csv', header=None)
sample_data = sample_data.dropna(how="any")
```

```python
sample_data.head()
```

```python
# function to extract only cities from location column
def split_record(info):
    ret_val = info.split(":")[1]
    return ret_val
```

```python
sample_data["Description"] = sample_data[3].map(split_record)
```
